```python

# Ensure the types are correct for merging
df_oecd['TIME_PERIOD'] = df_oecd['TIME_PERIOD'].astype(int)

# Separating into two tables...

df_oecd_policy_stringencies = df_oecd[df_oecd['Measure'].str.contains('Policy stringency')]

df_oecd_policy_counts = df_oecd[df_oecd['Measure'].str.contains('Adopted policies')]

df_oecd_policy_stringencies_wide = df_oecd_policy_stringencies.pivot_table(
    index=['REF_AREA', 'TIME_PERIOD'], 
    columns='CLIM_ACT_POL', 
    values='OBS_VALUE'
).reset_index() 

df_oecd_policy_counts_wide = df_oecd_policy_counts.pivot_table(
    index=['REF_AREA', 'TIME_PERIOD'], 
    columns='CLIM_ACT_POL', 
    values='OBS_VALUE'
).reset_index()

print(df_oecd_policy_stringencies_wide.head())
print(df_oecd_policy_counts_wide.head())

# Pivot the data so policies become columns
df_oecd_wide = df_oecd.pivot_table(
    index=['REF_AREA', 'TIME_PERIOD'], 
    columns='CLIM_ACT_POL', 
    values='OBS_VALUE'
).reset_index()

# Rename to match other datasets
df_oecd_wide = df_oecd_wide.rename(columns={'REF_AREA': 'Country_code_A3', 'TIME_PERIOD': 'Year'})
df_oecd_policy_counts_wide = df_oecd_policy_counts_wide.rename(columns={'REF_AREA': 'Country_code_A3', 'TIME_PERIOD': 'Year'})
df_oecd_policy_stringencies_wide = df_oecd_policy_stringencies_wide.rename(columns={'REF_AREA': 'Country_code_A3', 'TIME_PERIOD': 'Year'})

# NOTE: the value is sometimes a stringency score and sometimes a policy count so need to account for this.
```
