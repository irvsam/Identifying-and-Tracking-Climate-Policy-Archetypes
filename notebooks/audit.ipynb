```python
import wbgapi as wb
import pandas as pd

# 1. Define 19 G20 sovereign nations
countries = ['ARG', 'AUS', 'BRA', 'CAN', 'CHN', 'FRA', 'DEU', 'IND', 
             'IDN', 'ITA', 'JPN', 'MEX', 'RUS', 'SAU', 'ZAF', 'KOR', 
             'TUR', 'GBR', 'USA']

# 2. Define indicators we are interested in TODO: add more indicators related to climate policy, energy transition, etc.
indicators = {
    'EG.USE.PCAP.KG.OE': 'energy_use_per_capita',
    'NY.GDP.MKTP.KD': 'gdp_constant_usd',
    'EG.ELC.RNWX.ZS': 'renewable_electricity_pct',
    'EG.ELC.COAL.ZS': 'coal_electricity_pct',
    'NY.GDP.PCAP.KD': 'gdp_per_capita_constant'
}

# 3. Fetch the data
# wbgapi automatically returns a clean pandas DataFrame
df_wb = wb.data.DataFrame(indicators.keys(), countries, time=range(2015, 2026), labels=False)

# Rename the columns to our readable names
df_wb = df_wb.rename(columns=indicators)

# Set display format to show 2 decimal places and avoid scientific notation
pd.options.display.float_format = '{:,.2f}'.format

# Now print data (head is enough to check if it looks correct)
print(df_wb.head())
```
