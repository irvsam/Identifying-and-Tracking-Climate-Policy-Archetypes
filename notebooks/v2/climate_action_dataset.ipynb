```python
import pandas as pd


# Combing all datasets

gain = pd.read_csv('../../data/processed/gain_cleaned.csv')
wb = pd.read_csv('../../data/processed/world_bank_indicators_targeted.csv')
oecd = pd.read_csv('../../data/processed/oecd_policy_stringencies_wide.csv')

# Merge on country_code and year
# Rename oecd country code column and year
oecd.rename(columns={'Country_code_A3': 'country_code'}, inplace=True)
oecd.rename(columns={'Year': 'year'}, inplace=True)

# Merge GAIN and World Bank first
df_merged = pd.merge(gain, wb, on=['country_code', 'year'], how='outer')    
# Then merge the result with OECD
df_merged = pd.merge(df_merged, oecd, on=['country_code', 'year'], how='outer')

# Filter out any years after 2023
df_merged = df_merged[df_merged['year'] <= 2023]


print(df_merged.head())



# Save the merged dataset for modeling
df_merged.to_csv('../../data/processed/merged_dataset.csv', index=False)
```
