```python
import wbgapi as wb
import pandas as pd
import sys
import os

# Get the absolute path of the directory containing this nb
current_dir = os.getcwd()

project_root = os.path.abspath(os.path.join(current_dir, '../..'))

if project_root not in sys.path:
    sys.path.append(project_root)

from src.constants import countries, years


# Load
df_nd = pd.read_csv('../../data/raw/gain.csv')

countries = countries
years = years

# Filter dataframe to only include specified countries
df_nd = df_nd[df_nd['ISO3'].isin(countries)]


```

```python
# Melt to long
df_nd_melted = df_nd.melt(id_vars=['ISO3', 'Name'], 
                          var_name='year', 
                          value_name='nd_gain_score')


# Clean up year and rename for merging
df_nd_melted['year'] = df_nd_melted['year'].astype(int)
# Only include years we care about
df_nd_melted = df_nd_melted[df_nd_melted['year'].isin(years)]
df_nd_melted.rename(columns={'ISO3': 'country_code'}, inplace=True)

print(df_nd_melted.head())

# Save the cleaned GAIN data for later merging
df_nd_melted.to_csv('../../data/processed/gain_cleaned.csv', index=False)
```
